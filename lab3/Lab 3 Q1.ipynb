```python
# Importing relevant libraries/modules
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the dataset
digits = datasets.load_digits()

# Split into x and y data
x = digits.data
y = digits.target

# Test size 0.2 means 20% test and 80% training
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
test_scores = []
train_scores = []

for i in range (1, 9):
    # Create the classifier and use i nearest neighbours
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    test_scores.append(knn.score(x_test, y_test))
    train_scores.append(knn.score(x_train, y_train))
    
print(test_scores)
print(train_scores)
```

```python
plt.plot(range(1,9), train_scores, label = "Train Scores")
plt.plot(range(1,9), test_scores, label = "Test Scores")
plt.legend()
plt.xlabel("Number of neighbours")
plt.ylabel("Accuracy")
plt.show()
```

From the graph it is seen that k=1 has the best training accuracy and the equal best test accuracy. But using k=1 is overfitting the data points. Therefore it is concluded that k=3 is the best kNN number for both training and testing accuracy.

```python
# Creating the final kNN variable to be used to comparatively graph later on
knn_final = KNeighborsClassifier(n_neighbors=3)
knn_final.fit(x_train, y_train)
```

```python
# Question 2
from sklearn.metrics import classification_report
from sklearn.svm import SVC

svc = SVC(kernel="linear")
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
print(classification_report(y_test, y_pred))
```

```python
svc_poly = SVC(kernel="poly")
svc_poly.fit(x_train, y_train)
y_pred_poly = svc_poly.predict(x_test)
print(classification_report(y_test, y_pred_poly))
```

```python
svc_rbf = SVC(kernel="rbf")
svc_rbf.fit(x_train, y_train)
y_pred_rbf = svc_rbf.predict(x_test)
print(classification_report(y_test, y_pred_rbf))
```

From the f1-scores using different types of kernels it is shown that the poly and rbf have the same mean f1-score across each digit tested against. The linear kernel had the lowest f1-score mean, but by only a small margin. In this case either the poly or rbf kernel should be used.

```python
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
import seaborn as sns

bayes = GaussianNB()
bayes.fit(x_train, y_train)
y_pred_bayes = bayes.predict(x_test)

accuracy_matrix = confusion_matrix(y_test, y_pred_bayes)
plt.figure(figsize=(10,8))
sns.heatmap(accuracy_matrix, annot=True)
plt.xlabel("Predicted Digit")
plt.ylabel("Actual Digit")
plt.show()
```

```python
data1 = knn_final.score(x_test, y_test)
data2 = svc_rbf.score(x_test, y_test)
data3 = bayes.score(x_test, y_test)

graphData = {'KNN':data1, 'SVC':data2, 'Bayes':data3}

names = list(graphData.keys())
vals = list(graphData.values())

plt.bar(range(len(graphData)), vals, tick_label=names)
plt.ylabel('Accuracy')
plt.ylabel('Classifier')
plt.title('Various Classifier Accuracies for MNIST Data Set')
plt.show()

```
