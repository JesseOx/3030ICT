```python
# Importing relevant libraries/modules
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading the dataset
digits = datasets.load_digits()

# Split into x and y data
x = digits.data
y = digits.target

# Test size 0.2 means 20% test and 80% training
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
test_scores = []
train_scores = []

for i in range (1, 9):
    # Create the classifier and use i nearest neighbours
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    test_scores.append(knn.score(x_test, y_test))
    train_scores.append(knn.score(x_train, y_train))
    
print(test_scores)
print(train_scores)
```

```python
plt.plot(range(1,9), train_scores, label = "Train Scores")
plt.plot(range(1,9), test_scores, label = "Test Scores")
plt.xlabel("Number of neighbours")
plt.ylabel("Accuracy")
plt.show()
```
